### üìñ Teor√≠a: Anatom√≠a de una gr√°fica

Una gr√°fica tiene estos elementos b√°sicos:
- **Eje X (horizontal)**: La variable independiente
- **Eje Y (vertical)**: La variable dependiente
- **L√≠nea/Puntos**: Los datos representados
- **T√≠tulo**: Qu√© muestra la gr√°fica
- **Etiquetas**: Nombres de los ejes
- **Leyenda**: Identificaci√≥n de m√∫ltiples series

### üß™ EJEMPLO RESUELTO: Gr√°fica b√°sica

```python
# Datos de ejemplo: temperatura durante una semana
dias = [1, 2, 3, 4, 5, 6, 7]
temperaturas = [18, 20, 22, 21, 23, 25, 24]

# Crear la gr√°fica
plt.plot(dias, temperaturas)
plt.xlabel('D√≠a de la semana')
plt.ylabel('Temperatura (¬∞C)')
plt.title('Temperaturas de la Semana')
plt.grid(True, alpha=0.3)  # A√±adir cuadr√≠cula
plt.show()

print("‚úÖ Primera gr√°fica creada")
```

## üé® PASO 3: Personalizar gr√°ficas

### üìñ Teor√≠a: Hacer gr√°ficas m√°s informativas

Puedes personalizar:
- **Color**: `color='red'`, `color='#FF5733'`
- **Estilo de l√≠nea**: `linestyle='-'` (s√≥lida), `'--'` (punteada), `'-.'`, `':'`
- **Marcadores**: `marker='o'` (c√≠rculos), `'s'` (cuadrados), `'^'` (tri√°ngulos)
- **Grosor**: `linewidth=2`
- **Transparencia**: `alpha=0.7` (0=transparente, 1=opaco)

### üß™ EJEMPLO RESUELTO: Gr√°fica personalizada

```python
# Datos de dos productos
meses = np.arange(1, 13)  # 12 meses
producto_a = [100, 120, 150, 140, 160, 180, 170, 190, 200, 210, 220, 240]
producto_b = [80, 90, 95, 100, 110, 120, 130, 140, 150, 160, 170, 180]

plt.figure(figsize=(12, 6))

# Gr√°fica del Producto A
plt.plot(meses, producto_a,
         color='blue',          # Color azul
         linestyle='-',         # L√≠nea s√≥lida
         marker='o',            # Marcadores circulares
         linewidth=2,           # Grosor de l√≠nea
         label='Producto A')    # Etiqueta para leyenda

# Gr√°fica del Producto B
plt.plot(meses, producto_b,
         color='red',
         linestyle='--',        # L√≠nea punteada
         marker='s',            # Marcadores cuadrados
         linewidth=2,
         label='Producto B')

plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (unidades)', fontsize=12)
plt.title('Comparaci√≥n de Ventas Mensuales', fontsize=14, fontweight='bold')
plt.legend()  # Mostrar leyenda
plt.grid(True, alpha=0.3)
plt.tight_layout()  # Ajustar espaciado
plt.show()
```

## üß† PASO 5: Visualizar funciones de activaci√≥n

### üìñ Teor√≠a: Funciones de activaci√≥n en redes neuronales

Las **funciones de activaci√≥n** determinan c√≥mo las neuronas procesan la informaci√≥n:

1. **ReLU (Rectified Linear Unit)**: `f(x) = max(0, x)`
   - Usada en capas ocultas
   - Simple y eficiente

2. **Sigmoid**: `f(x) = 1 / (1 + e^(-x))`
   - Salida entre 0 y 1
   - Usada en clasificaci√≥n binaria

3. **Tanh**: `f(x) = (e^x - e^(-x)) / (e^x + e^(-x))`
   - Salida entre -1 y 1
   - Versi√≥n centrada de sigmoid

### üß™ EJEMPLO RESUELTO: Graficar funciones de activaci√≥n

```python
# Definir las funciones de activaci√≥n
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

# Generar valores de x
x = np.linspace(-5, 5, 100)

# Calcular valores de y para cada funci√≥n
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Crear la visualizaci√≥n
plt.figure(figsize=(15, 5))

# ReLU
plt.subplot(1, 3, 1)
plt.plot(x, y_relu, color='blue', linewidth=2)
plt.title('ReLU', fontsize=14, fontweight='bold')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='--', linewidth=0.5)

# Sigmoid
plt.subplot(1, 3, 2)
plt.plot(x, y_sigmoid, color='green', linewidth=2)
plt.title('Sigmoid', fontsize=14, fontweight='bold')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0.5, color='r', linestyle='--', linewidth=0.5, alpha=0.5)
plt.axvline(x=0, color='k', linestyle='--', linewidth=0.5)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, color='red', linewidth=2)
plt.title('Tanh', fontsize=14, fontweight='bold')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()

print("\nüìä CARACTER√çSTICAS DE CADA FUNCI√ìN:")
print("="*60)
print("ReLU: Rango [0, ‚àû) | Simple y r√°pida")
print("Sigmoid: Rango [0, 1] | Para clasificaci√≥n binaria")
print("Tanh: Rango [-1, 1] | Centrada en cero")
```

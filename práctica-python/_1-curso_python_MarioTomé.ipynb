## Lecci√≥n 5: Tu Primer Bucle - Mostrando una Lista

Un bucle `for` es como decir "para cada cosa en esta lista, haz algo". Es √∫til cuando quieres hacer la misma acci√≥n con cada elemento de una lista, sin tener que escribir el c√≥digo muchas veces.

Piensa en esto como si tuvieras una fila de cajas y quisieras abrir cada una. En vez de escribir "abre la primera caja, abre la segunda caja, abre la tercera caja...", simplemente dices "para cada caja en la fila, √°brela".

````python
# El bucle for m√°s simple

# Tienes una lista
numeros = [1, 2, 3]

# Quieres mostrar cada n√∫mero
for numero in numeros:
    print(numero)
````

Lo que est√° pasando es: Python va n√∫mero por n√∫mero en la lista, y en cada vuelta muestra ese n√∫mero. Observa c√≥mo el c√≥digo dentro del bucle (el print) est√° **indentado** (desplazado a la derecha). Esto es importante en Python para indicar qu√© c√≥digo pertenece al bucle.

````python
# Bucle con nombres

nombres = ["Ana", "Luis", "Carmen"]

for nombre in nombres:
    print("Hola", nombre)
````

A veces no tienes una lista preparada, pero quieres repetir algo un n√∫mero espec√≠fico de veces. Para eso usamos `range()`.

## Lecci√≥n 8: Guardando Resultados en una Lista Nueva

Cuando haces c√°lculos en un bucle, a menudo quieres guardar todos los resultados para usarlos despu√©s. Para eso, creas una lista vac√≠a antes del bucle y vas a√±adiendo resultados con `.append()`.

Este es un patr√≥n muy com√∫n en programaci√≥n y lo usar√°s constantemente en Machine Learning.

````python
# Guardando resultados en una nueva lista

# N√∫meros originales
numeros = [1, 2, 3]

# Lista vac√≠a para guardar resultados
dobles = []

# Para cada n√∫mero, calcula su doble y gu√°rdalo
for numero in numeros:
    doble = numero * 2
    dobles.append(doble)

# Al final, muestra todos los resultados
print("N√∫meros originales:", numeros)
print("Los dobles:", dobles)
````

````python
# Otro ejemplo paso a paso

numeros = [5, 10, 15]
resultados = []  # Lista vac√≠a al principio

print("Procesando cada n√∫mero...")
for numero in numeros:
    # Suma 5 a cada n√∫mero
    nuevo_numero = numero + 5
    print(f"  {numero} + 5 = {nuevo_numero}")
    # Gu√°rdalo en la lista de resultados
    resultados.append(nuevo_numero)

print("\nLista final de resultados:", resultados)
````

## Lecci√≥n 11: Bucles con Decisiones Dentro

## Lecci√≥n 13: Funciones que Reciben Informaci√≥n

Las funciones son m√°s √∫tiles cuando pueden trabajar con diferentes valores. Para eso, les pasas informaci√≥n entre par√©ntesis. Esta informaci√≥n se llama "par√°metros" o "argumentos".

````python
# Funci√≥n con un par√°metro

def saludar(nombre):
    print("Hola", nombre)

saludar("Ana")   # Mostrar√°: Hola Ana
saludar("Luis")  # Mostrar√°: Hola Luis
````

````python
# Funci√≥n con dos par√°metros

def sumar(a, b):
    resultado = a + b
    print("La suma de", a, "y", b, "es:", resultado)

sumar(5, 3)   # Mostrar√°: La suma de 5 y 3 es: 8
sumar(10, 20) # Mostrar√°: La suma de 10 y 20 es: 30
````

# üéì Proyecto Final: Calculando Promedios

Ahora vamos a juntar todo lo que aprendiste en un proyecto simple pero completo. Vamos a calcular el promedio de unas notas paso a paso.

Este es el tipo de c√°lculo que har√°s constantemente en Machine Learning para evaluar qu√© tan bien funciona un modelo.

````python
# Proyecto Final: Calculadora de Promedios

print("="*50)
print("CALCULADORA DE PROMEDIOS")
print("="*50)

# Paso 1: Tenemos las notas
notas = [7, 8, 6, 9, 7]
print("\nNotas del estudiante:", notas)

# Paso 2: Creamos una funci√≥n para sumar todas las notas
def sumar_lista(lista):
    total = 0  # Empezamos con cero

    # Sumamos cada nota al total
    for numero in lista:
        total = total + numero

    return total  # Devolvemos el resultado

# Paso 3: Probamos la funci√≥n
suma_total = sumar_lista(notas)
print("La suma total de notas es:", suma_total)

# Paso 4: Calculamos el promedio
def calcular_promedio(lista):
    # Sumamos todos los n√∫meros
    total = sumar_lista(lista)

    # Dividimos entre cu√°ntos n√∫meros hay
    cantidad = len(lista)  # len() cuenta cu√°ntos elementos hay
    promedio = total / cantidad

    return promedio

# Usamos la funci√≥n
promedio_final = calcular_promedio(notas)
print("El promedio es:", promedio_final)

# Decidimos si aprob√≥ o no
print("\nEvaluaci√≥n:")
if promedio_final >= 5:
    print("‚úÖ ¬°Aprobado!")
else:
    print("‚ùå Suspendido")
````
